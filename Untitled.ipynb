```python
import fitz
import json
import re
import os
pdf_path=r""
fileName="summary records sample.pdf"
doc=fitz.open(os.path.join(pdf_path,fileName))
all_text=[]
for i in range(len(doc)):
    page=doc[i]
    text=page.get_text()
    all_text.append(text)
with open(f"all_context{fileName.split('.pdf')[0]}.json","w+")as f:
    json.dump(all_text,f,indent=4)
len(all_text)
```
